```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data from your CSV file
data = pd.read_csv('kepler-cpu-seconds-example.csv', delimiter=';')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the Seaborn style and pastel color palette
sns.set_theme(style="whitegrid", palette="pastel")

# Create the figure and axes
fig, ax = plt.subplots(figsize=(12, 6))

# Plot "Pi (CPU)" and "All (CPU)" data on the primary y-axis
sns.lineplot(data=data, x='Zeit', y='Pi (CPU)', label='Pi (CPU)', linestyle=':', linewidth=2, ax=ax)
sns.lineplot(data=data, x='Zeit', y='All (CPU)', label='All (CPU)', linestyle=':', linewidth=2, ax=ax)

# Fill under the primary y-axis curves for clarity
ax.fill_between(data['Zeit'], data['Pi (CPU)'], alpha=0.2)
ax.fill_between(data['Zeit'], data['All (CPU)'], alpha=0.2)

# Set primary y-axis limits, labels, and legend
ax.set_ylim(0, 1000)
ax.set_xlim(data['Zeit'].min(), data['Zeit'].max())
ax.set_xlabel('Time in seconds', fontsize=12)
ax.set_ylabel('CPU seconds', fontsize=12)
ax.grid(False)


# add horizontal line at 450 cpu seconds "target 50% cpu utilization"
ax.axhline(y=470, color='r', linestyle='--', label='target 50% CPU Utilization')

ax.legend(loc='upper left', fontsize=10, frameon=True)

# Create a secondary y-axis for the "All (Kepler)" and "Pi (Kepler)" data
ax_right = ax.twinx()

# Plot "All (Kepler)" and "Pi (Kepler)" data on the secondary y-axis
sns.lineplot(data=data, x='Zeit', y='All (Kepler)', label='All (Watt)', linestyle='-', linewidth=2, ax=ax_right)
sns.lineplot(data=data, x='Zeit', y='Pi (Kepler)', label='Pi (Watt)', linestyle='-', linewidth=2, ax=ax_right)

# Fill under the secondary y-axis curves for clarity
#ax_right.fill_between(data['Zeit'], data['All (Kepler)'], alpha=0.2)
#ax_right.fill_between(data['Zeit'], data['Pi (Kepler)'], alpha=0.2)

# Set secondary y-axis limits, labels, and legend
ax_right.set_ylim(230, 300)
ax_right.set_ylabel('Watt', fontsize=12)
ax_right.legend(loc='upper right', fontsize=10, frameon=True)



# Adjust layout for clarity
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Set the Seaborn style and pastel color palette
sns.set_theme(style="whitegrid", palette="pastel")

# Create the plot
plt.figure(figsize=(12, 6))

# Plot "Pi (CPU)" and "All (CPU)" data with fill
sns.lineplot(data=data, x='Zeit', y='Pi (CPU)', label='Pi (CPU)', linestyle=':', linewidth=2)
sns.lineplot(data=data, x='Zeit', y='All (CPU)', label='All (CPU)', linestyle=':', linewidth=2)
sns.set_palette("pastel")

# Fill between lines for better visualization
plt.fill_between(data['Zeit'], data['Pi (CPU)'], alpha=0.3)
plt.fill_between(data['Zeit'], data['All (CPU)'], alpha=0.3)

# Set limits and labels
plt.ylim(0, 1000)
plt.xlim(data['Zeit'].min(), data['Zeit'].max())
plt.xlabel('Time in seconds', fontsize=12)
plt.ylabel('CPU seconds', fontsize=12)

# Style and legend
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)

# add horizontal line at 450 cpu seconds "target 50% cpu utilization"
plt.axhline(y=470, color='r', linestyle='--', label='target 50% CPU Utilization')

plt.legend(loc='upper left', fontsize=10, frameon=True, fancybox=True, framealpha=0.8)

# Show the plot
plt.tight_layout()
plt.show()
```
