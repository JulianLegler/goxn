```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from your CSV file
data = pd.read_csv('kepler-cpu-seconds-example.csv', delimiter=';')
```

```python
# Plotting the data
fig, ax = plt.subplots(figsize=(12, 6))

# Plot "Pi" data on the primary y-axis

ax.plot(data['Zeit'], data['Pi (CPU)'], label='Pi (CPU)', color='green', linestyle=':', linewidth=2)
ax.plot(data['Zeit'], data['All (CPU)'], label='All (CPU)', color='blue', linestyle=':', linewidth=2)
# Set primary y-axis limits for "Pi" data
ax.set_ylim(0, 1500)
ax.set_xlabel('Time in seconds')
ax.set_ylabel('CPU seconds')

ax.set_xlim(data['Zeit'].min(), data['Zeit'].max())


# Create a secondary y-axis for the "All" data
ax_right = ax.twinx()

# Plot "All" data on the secondary y-axis
ax_right.plot(data['Zeit'], data['All (Kepler)'], label='All (Watt)', color='blue', linestyle='-', linewidth=2)
ax_right.plot(data['Zeit'], data['Pi (Kepler)'], label='Pi (Watt)', color='green', linestyle='-', linewidth=2)

# Set secondary y-axis limits for "All" data
ax_right.set_ylim(230, 300)
ax_right.set_ylabel('Watt')

# Adding legends for both axes
ax.legend(loc='upper left')
ax_right.legend(loc='upper right')

# Display the plot
plt.show()
```

```python
# Plotting the data
fig, ax = plt.subplots(figsize=(6, 6))

# Plot "Pi" data on the primary y-axis

ax.plot(data['Zeit'], data['Pi (CPU)'], label='Pi (CPU)', color='green', linestyle='-', linewidth=2)
ax.plot(data['Zeit'], data['All (CPU)'], label='All (CPU)', color='blue', linestyle='-', linewidth=2)
# Set primary y-axis limits for "Pi" data
ax.set_ylim(0, 1500)
ax.set_xlabel('Time in seconds', fontsize=12)
ax.set_ylabel('CPU seconds', fontsize=12)

ax.set_xlim(data['Zeit'].min(), data['Zeit'].max())


# Style stuff
ax.tick_params(axis='x', rotation=45, labelsize=10)
ax.tick_params(axis='y', labelsize=10)
ax.grid(True, which='both', color='gray', linestyle='--', linewidth=0.5, alpha=0.7)
ax.legend(loc='upper left', fontsize=10, frameon=True, fancybox=True, framealpha=0.8)

# Display the plot
plt.show()
```
